```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Baca gambar grayscale
# Gunakan variabel file_name yang sudah ada dari cell sebelumnya
img = cv2.imread(file_name, 0)

# Negatif
negatif = 255 - img

# Logaritmik
c = 255 / np.log(1 + np.max(img))
log_img = c * np.log(1 + img)
log_img = np.array(log_img, dtype=np.uint8)

# Histogram
plt.figure(figsize=(10,8))

plt.subplot(2,4,1)
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(2,4,2)
plt.imshow(negatif, cmap='gray')
plt.title('Negatif')
plt.axis('off')

plt.subplot(2,4,3)
plt.imshow(log_img, cmap='gray')
plt.title('Logaritmik')
plt.axis('off')

plt.subplot(2,4,5)
plt.hist(img.ravel(), 256, [0,256])
plt.title('Hist Original')

plt.subplot(2,4,6)
plt.hist(negatif.ravel(), 256, [0,256])
plt.title('Hist Negatif')

plt.subplot(2,4,7)
plt.hist(log_img.ravel(), 256, [0,256])
plt.title('Hist Log')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Baca gambar grayscale menggunakan variabel file_name
img = cv2.imread(file_name, 0)

# Tambahkan pengecekan jika gambar tidak ditemukan
if img is None:
    print(f"Error: '{file_name}' not found. Please ensure the image file is in the correct directory.")
    # Membuat gambar dummy untuk demonstrasi agar kode selanjutnya tidak error
    img = np.zeros((200, 300), dtype=np.uint8) # Contoh gambar hitam 200x300
    print("Using a dummy black image for demonstration.")

# Kernel
lpf = (1/9)*np.ones((3,3))
hpf = np.array([[-1,-1,-1],
                [-1, 8,-1],
                [-1,-1,-1]])
bpf = np.array([[0,-1,0],
                [-1,5,-1],
                [0,-1,0]])

# Konvolusi
img_lpf = cv2.filter2D(img, -1, lpf)
img_hpf = cv2.filter2D(img, -1, hpf)
img_bpf = cv2.filter2D(img, -1, bpf)

# Tampilkan
plt.figure(figsize=(10,6))

plt.subplot(2,2,1)
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(img_lpf, cmap='gray')
plt.title('LPF (Blur)')
plt.axis('off')

plt.subplot(2,2,3)
plt.imshow(img_hpf, cmap='gray')
plt.title('HPF (Edge)')
plt.axis('off')

plt.subplot(2,2,4)
plt.imshow(img_bpf, cmap='gray')
plt.title('BPF (Sharpen)')
plt.axis('off')

plt.show()
```
